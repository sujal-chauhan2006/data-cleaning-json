```python
import json
import datetime as datetime
```

```python
def data_read():
  global data
  data = [
    {"id": 1, "name": "Alice", "email": "alice@example.com", "age": 28},
    {"id": 2, "name": "Bob", "email": "bob@example.com", "age": 17},
    {"id": 3, "name": "Coral", "email": "carol@example.com", "age": 120}
  ]

  with open("users.json", "w") as file:
      json.dump(data, file)
  try:
    with open("users.json", "r",encoding="utf-8") as file:
      data = json.load(file)

  except FileNotFoundError:
    print("File Not Found")
  except json.JSONDecodeError:
    print("Invalide Json File")
  except Exception as e:
    print("Unknown Error",e)
```

```python
def validate(data):
  for i, user in enumerate(data):


    required_fields = ["id","name","email","age"]

    for fields in required_fields:
      if fields not in user:
        print(f"Invalid {fields}")
        return
    # type Validation

    if type(user["id"]) != int:
      print("Invalid id type")

    if type(user["name"]) != str:
      print("Invalid name type")

    if not isinstance(user["email"], str) or "@" not in user["email"]:
      print("Invalid email")

    if type(user["age"]) != int:
      print("Invalid age type")

    # Age range validation
    if not (0 <= user["age"] <= 150):
      print("Invalid Age")

```

```python
def Data_Transform(data):
  def age_group(age):
    for user in data:
      if user['age'] < 18:
        user['status'] = "Minor"
      elif user['age'] < 65:
        user['status'] = "Adult"
      elif user['age'] > 65:
        user['status'] = "Senior"

    with open("users.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
  age_group(data)

  def validated_at(data):
    for user in data:
      user["validated_at"] = datetime.datetime.now().strftime("%Y-%m-%dT%H:%M:%S")
  validated_at(data)

  with open("users_processed.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)

  def summary_statistics(data):
    total_users = len(data)
    total_age = 0
    for user in data:
      total_age += user['age']
    avg_age = total_age / total_users

    def age_dist(data):
      age_dist = {'Minor':0,
                  'Adult':0,
                  'Senior':0
                  }
      for user in data:
        if user['age'] <= 18:
          age_dist['Minor'] += 1
        elif user['age'] <= 65:
          age_dist['Adult'] += 1
        elif user['age'] > 65:
          age_dist['Senior'] += 1
      return age_dist

    # print(f'Total Users: {total_users}')
    # print(f'Average Age: {avg_age}')
    # print(f'Age Distribution:{age_dist(data)}')

    summary = {
    "total_users": total_users,
    "average_age": avg_age,
    "age_distribution": age_dist(data)
    }

    with open("summary.json", "w", encoding="utf-8") as f:
      json.dump(summary, f, indent=2)

  summary_statistics(data)


```
